```python
import cv2 as cv
import matplotlib.pyplot as plt

# Load the image
input_image_path = '/Users/kavindujayathissa/Desktop/Academics/Semester 05/EN3160 Image Processing/Assignment A01/a1images/brain_proton_density_slice.png'
input_image = cv.imread(input_image_path, cv.IMREAD_GRAYSCALE)

# Display the image with a scale
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray', aspect='auto')  # Display in grayscale
plt.colorbar(label='Pixel Intensity')  # Add a color scale
plt.title('Image with Pixel Intensity Scale')
plt.xlabel('X Coordinate (pixels)')
plt.ylabel('Y Coordinate (pixels)')

# Set the x and y ticks to correspond to the pixel values
plt.xticks(range(0, input_image.shape[1], 50))  # X ticks every 50 pixels
plt.yticks(range(0, input_image.shape[0], 50))  # Y ticks every 50 pixels

# Show the image plot
plt.subplot(1, 2, 2)

# Create a histogram of the pixel intensities
plt.hist(input_image.ravel(), bins=256, range=[0, 256], color='gray', alpha=0.7)
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid()

plt.tight_layout()
plt.show()  # Show both the image and the histogram
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv.imread('/Users/kavindujayathissa/Desktop/Academics/Semester 05/EN3160 Image Processing/Assignment A01/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)

# Define the intensity transformation
t = np.zeros(256, dtype=np.uint8)  
t[100:181] = 255  # Set values between 100 and 180 to 255
# Apply the transformation using the LUT function
output_image = cv.LUT(input_image, t)

# Plot the intensity transformation graph
plt.figure(figsize=(10, 5))
plt.plot(range(256), t, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')
plt.legend()
plt.grid()
plt.show()

# Plot original and enhanced images side by side
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')  # Hide axes

plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.imshow(output_image, cmap='gray')
plt.title('Gray Matter Enhanced Image')
plt.axis('off')  # Hide axes

plt.tight_layout()  # Adjust subplots to fit in the figure area
plt.show()  # Display the images
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os

# Load image path from an environment variable for flexibility
image_path = os.getenv('IMAGE_PATH', '/Users/kavindujayathissa/Desktop/Academics/Semester 05/EN3160 Image Processing/Assignment A01/a1images/brain_proton_density_slice.png')

# Read the input image
input_image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

# Define the intensity transformation: set 180-255 to 255, all others to 0
t = np.zeros(256, dtype=np.uint8)
t[180:256] = 255
# Create the output image using the LUT function
output_image = cv.LUT(input_image, t)

# Plot the intensity transformation graph
plt.figure(figsize=(10, 5))
plt.plot(range(256), t, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')
plt.legend()
plt.grid()
plt.show()

# Plot original and enhanced images side by side
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(output_image, cmap='gray')
plt.title('White Matter Enhanced Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
