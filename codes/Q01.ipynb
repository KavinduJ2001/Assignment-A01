```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv.imread('/Users/kavindujayathissa/Desktop/Academics/Semester 05/EN3160 Image Processing/Assignment A01/a1images/emma.jpg', cv.IMREAD_GRAYSCALE)  

t1 = np.linspace(0,50,51)
t2 = np.linspace(51,100,0)
t3 = np.linspace(100,255,100)
t4 = np.linspace(255,150,0)
t5 = np.linspace(151,255,105)
t = np.concatenate((t1,t2,t3,t4,t5), axis=0).astype('uint8')

# Create the output image
output_image = cv.LUT(input_image, t)

# Plot the intensity transformation graph
plt.plot(range(256), t, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')
plt.legend()


# Save or display the output image
cv.imwrite('Q1_output_emma.png', output_image)  
cv.imshow('Output Image', output_image)
cv.waitKey(1000) 
cv.destroyAllWindows()
```
