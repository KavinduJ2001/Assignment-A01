```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Read the input image
input_image = cv.imread('/Users/kavindujayathissa/Desktop/Academics/Semester 05/EN3160 Image Processing/Assignment A01/a1images/brain_proton_density_slice.png', cv.IMREAD_GRAYSCALE)  

t1 = np.linspace(0,25,150)
t2 = np.linspace(26,255,106)
t = np.concatenate((t1,t2), axis=0).astype(np.uint8)

# Create the output image
output_image = cv.LUT(input_image, t)

# Plot the intensity transformation graph
plt.plot(range(256), t, label='Intensity Transformation')
plt.xlabel('Input Intensity')
plt.ylabel('Output Intensity')
plt.title('Intensity Transformation Graph')

# Save the plot as an image
plt.savefig('intensity_transformation_graph_Q2_b.png')


# Save or display the output image
cv.imwrite('Q2_output_gray_matter.jpg', output_image)  # Save the output image
cv.imshow('Output Image', output_image)
cv.waitKey(1000)
cv.destroyAllWindows()
```
